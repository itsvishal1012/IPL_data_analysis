```python
# Fill string/categorical missing with "None"
df['wicket_type'].fillna("None", inplace=True)
df['player_dismissed'].fillna("None", inplace=True)
df['fielder'].fillna("None", inplace=True)
```

```python
# Function to detect outliers using IQR
def detect_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"Outliers in {col}: {len(outliers)}")

for col in ['runs_of_bat', 'extras', 'wide', 'legbyes', 'byes', 'noballs']:
    detect_outliers(col)
```

```python
# Define numeric columns you want to check for outliers
numeric_cols = ['runs_of_bat', 'extras', 'wide', 'legbyes', 'byes', 'noballs']

# Create boxplots for all numeric features
plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.violinplot(x=df[col], inner='point', color='lightgreen')
    plt.title(f'Violin Plot of {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import zscore

# Calculate Z-scores
z_scores = df[numeric_cols].apply(zscore)

# Visualize Z-scores
plt.figure(figsize=(14, 8))
sns.boxplot(data=z_scores)
plt.axhline(y=3, color='r', linestyle='--', label='Z=3')
plt.axhline(y=-3, color='r', linestyle='--')
plt.title("Z-score Based Outlier Detection")
plt.legend()
plt.show()
```

```python
# Runs vs Extras
sns.scatterplot(data=df, x="runs_of_bat", y="extras")
plt.title("Runs off Bat vs Extras")
plt.show()
```

```python
# Top Run Scorers
top_strikers = df.groupby("striker")["runs_of_bat"].sum().sort_values(ascending=False).head(10)
top_strikers.plot(kind='bar', title='Top 10 Run Scorers')
plt.ylabel("Runs")
plt.xticks(rotation=90)
plt.show()
```

```python
# Top Bowlers by Wickets
wickets_df = df[df["player_dismissed"].notnull()]
top_bowlers = wickets_df["bowler"].value_counts().head(10)
top_bowlers.plot(kind='bar', title="Top 10 Bowlers by Wickets")
plt.ylabel("Wickets")
plt.xticks(rotation=90)
plt.show()
```

```python
# Dismissal Types
dismissal_counts = df["wicket_type"].value_counts()
dismissal_counts.plot(kind="bar", title="Dismissal Types")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Overs vs Runs Trend
over_runs = df.groupby("over")["runs_of_bat"].mean()
over_runs.plot(kind='line', marker='o', title='Average Runs Per Over')
plt.xlabel("Over")
plt.ylabel("Average Runs")
plt.show()
```

```python
# Extras Breakdown
extras_cols = ['wide', 'legbyes', 'byes', 'noballs']
df[extras_cols].sum().plot(kind="bar", title="Extras Breakdown")
plt.ylabel("Total Extras")
plt.xticks(rotation=0)
plt.show()
```

```python
# Batting Team Performance
batting_perf = df.groupby("batting_team")["runs_of_bat"].sum().sort_values(ascending=False)
batting_perf.plot(kind="bar", title="Total Runs by Team")
plt.ylabel("Runs")
plt.xticks(rotation=0)
plt.show()
```

```python
# Pie Chart: Team Contribution to Total Runs

team_runs = df.groupby("batting_team")["runs_of_bat"].sum()
plt.figure(figsize=(10, 10))
plt.pie(team_runs, labels=team_runs.index, autopct="%1.1f%%", startangle=90, colors=sns.color_palette("Set3"))
plt.title("Team Contribution to Total Runs")
plt.axis("equal")
plt.show()
```

```python
# Count Plot: Top 10 Bowlers by Wickets (Seaborn)

top_bowlers = df[df["player_dismissed"].notnull()]["bowler"].value_counts().head(10)
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette="magma")
plt.title("Top 10 Bowlers by Wickets")
plt.xlabel("Wickets")
plt.ylabel("Bowler")
plt.show()
```

```python
# Stacked Bar Chart: Runs vs Extras by Team

team_stats = df.groupby("batting_team")[["runs_of_bat", "extras"]].sum()
team_stats.sort_values("runs_of_bat", ascending=False).plot(kind="bar", stacked=True, color=["#1f77b4", "#ff7f0e"])
plt.title("Runs vs Extras by Team")
plt.ylabel("Total Runs")
plt.xticks(rotation=90)
plt.legend(["Runs of Bat", "Extras"])
plt.show()
```
